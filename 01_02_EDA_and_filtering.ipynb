```python
derma_meta = pd.read_csv('/content/drive/MyDrive/dermatology-research/data/raw/dermaconIN/METADATA/Skin_Metadata-1.csv')

print("=== DERMACONIN METADATA ===")
print(derma_meta.head())
print("\nColumns:", derma_meta.columns.tolist())
print("\nShape:", derma_meta.shape)
```

```python
# Fitzpatrick distribution
print("=== FITZPATRICK DISTRIBUTION ===")
print(derma_meta['Fitzpatrick'].value_counts())

print("\n=== MAIN DISEASE CATEGORIES ===")
print(derma_meta['Main_class'].value_counts())

print("\n=== FST V AND VI ONLY ===")
dark = derma_meta[derma_meta['Fitzpatrick'].isin(['FST 5', 'FST 6'])]
print("Total FST V-VI images:", len(dark))
print("\nDisease distribution:")
print(dark['Main_class'].value_counts())
```

```python
print("=== GRADABILITY CHECK ===")
dark = derma_meta[derma_meta['Fitzpatrick'].isin(['FST 5', 'FST 6'])]
print(dark['Gradability'].value_counts())

print("\n=== CONFIDENCE CHECK ===")
print(dark['Confidence'].value_counts())
```

```python
# Filter final clean dark skin dataset
dark_clean = derma_meta[
    (derma_meta['Fitzpatrick'].isin(['FST 5', 'FST 6'])) &
    (derma_meta['Confidence'] >= 4) &
    (derma_meta['Main_class'].isin([
        'Infectious Disorders',
        'Inflammatory Disorders',
        'Pigmentary Disorders'
    ]))
]

print("Final clean dark skin dataset size:", len(dark_clean))
print("\nDisease distribution:")
print(dark_clean['Main_class'].value_counts())
print("\nFST distribution:")
print(dark_clean['Fitzpatrick'].value_counts())
```

```python
# Save filtered metadata to CSV
output_path = '/content/drive/MyDrive/dermatology-research/data/filtered/fst_v_vi_only/dark_skin_metadata.csv'

dark_clean.to_csv(output_path, index=False)
print("Saved!", len(dark_clean), "images metadata saved")
```

```python
import os
import shutil
import pandas as pd

# Load your filtered metadata
dark_clean = pd.read_csv('/content/drive/MyDrive/dermatology-research/data/filtered/fst_v_vi_only/dark_skin_metadata.csv')

# Source and destination
src = '/content/drive/MyDrive/dermatology-research/data/raw/dermaconIN/DATASET'
dst = '/content/drive/MyDrive/dermatology-research/data/filtered/fst_v_vi_only/images'

# Create destination folder
os.makedirs(dst, exist_ok=True)

# Copy filtered images
copied = 0
not_found = 0

for img_name in dark_clean['Image_name']:
    src_path = os.path.join(src, img_name)
    dst_path = os.path.join(dst, img_name)

    if os.path.exists(src_path):
        shutil.copy2(src_path, dst_path)
        copied += 1
    else:
        not_found += 1

print(f"Copied: {copied}")
print(f"Not found: {not_found}")
print(f"Total expected: {len(dark_clean)}")
```
